9. Agrupar por año, seleccionar los ultimos 100 años y contar los casos

```python
df_year = df.groupby("Year")["Name"].count().reset_index()
df_year.tail()
df_100 = df_year[df_year["Year"] > 1917].reset_index(drop = True)
df_100.rename(columns = {"Name": "Cases"}, inplace = True)
df_100.head()
df_100.set_index("Year").T
```
